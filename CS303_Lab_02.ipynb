 1.   **Convert from RGB Image To Gray-Scale Image**

      Scikit-Image library provides the **rgb2gray** function to convert a color (RGB) image to a grayscale image.

      1.   You have to import the data module and color first
      2.   then load the image by calling it from the data module
      3. Call the rgb2gray function from color module and put Input_image as a prameter




```python
from skimage import data , color
import matplotlib.pyplot as plt
rgb_image = data.chelsea()

#convert from rgb to gray
gray_image = color.rgb2gray(rgb_image)

fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].imshow(rgb_image)
ax[0].set_title(" Orginal RGB Image ")
ax[0].axis("off")
ax[1].imshow(gray_image , cmap="gray")
ax[1].set_title(" Gray Scale Image ")
ax[1].axis("off")
```

**2. Convert from RGB Image To Binary Image**


1. First , Convert image from rgb to gray scale
2. Then , Convert gray scale to Binary
3. To Convert step 2 , You have to **import filters module** to call **threshold _otsu function**

  **threshold _otsu ()** --> Using to separate pixels




```python
from skimage import filters
# Applay threshold on gray image to separate pixels into two groups
threshold_value = filters.threshold_otsu(gray_image)
binary_image = gray_image > threshold_value

fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].imshow(rgb_image)
ax[0].set_title(" Orginal RGB Image ")
ax[0].axis("off")
ax[1].imshow(binary_image , cmap="binary_r")
ax[1].set_title("Binary Image ")
ax[1].axis("off")

```

**3.Convert From Gray Image To Binary Image**

1. You have to **import filters module** to call **threshold _otsu function**

  **threshold _otsu ()** --> Using to separate pixels






```python

page_gray_image = data.page()
# Applay threshold on gray image to separate pixels into two groups
threshold_value = filters.threshold_otsu(page_gray_image)
binary_image = page_gray_image > threshold_value
# Represent images
fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].imshow( page_gray_image , cmap ="gray")
ax[0].set_title(" Orginal Gray Scale Image ")
ax[0].axis("off")
ax[1].imshow(binary_image , cmap="binary_r")
ax[1].set_title("Binary Image ")
ax[1].axis("off")
```

**4. Convert Gray Scale Image To RGB Image**

1.   Import  color , data modules
2.   Read gray image from data module
3. Using **gray2rgb ()** function to convert gray image to rgb image



```python
from skimage import data , color
#Read Gray Image
coins_gray_image= data.coins()
#Convert Gray Image To RGB Image
coins_rgb_image=color.gray2rgb(coins_gray_image)
#Represent Images
fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].imshow(coins_gray_image , cmap="gray")
ax[0].set_title(" Orginal Gray Image ")
ax[0].axis("off")
ax[1].imshow(coins_rgb_image )
ax[1].set_title("RGB Image ")
ax[1].axis("off")
```

**5. Convert From Binary Images To RGB Images**
1.   Import  color , data modules
2.   Read gray image from data module
3. First , convert Binary Image to Gray Scale image by Multiple pixels by 255

  0= 0.255 = 0

  1=1.255 =  255
3. Using **gray2rgb ()** function to convert gray image to rgb image

```python
from skimage import data , color
#Read Gray Image
horse_binary_image= data.horse()
#convert Binary image to gray image
horse_gray_image=horse_binary_image*255
#Convert Gray Image To RGB Image
horse_rgb_image=color.gray2rgb(horse_gray_image)
#Represent Images
fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].imshow(horse_binary_image , cmap="binary_r")
ax[0].set_title(" Orginal Binary Image ")
ax[0].axis("off")
ax[1].imshow(horse_rgb_image )
ax[1].set_title("RGB Image ")
ax[1].axis("off")
```

**Distances Between Pixels**

There are many image processing
applications that require measuring distances between pixels


1.   Import **scipy.spatial** libarary
2.   Import **distance** to measure distance between 2 pixels  
3. Present image and its Shape to can select pixels
4. use distance to select which way you want to measure the distance



```python
from skimage import data
import matplotlib.pyplot as plt
from scipy.spatial import distance
image= data.clock()
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
print (image)
print(f" \n Image Shape : {image.shape}")
```

**Measure Distance between point 1 and Point 2 using Euclidean distance**

```python
# Select pixels
point1 = (100, 150)
point2 = (110, 50)
#Measure Distance between point 1 and Point 2 using Euclidean distance
euclidean_dist = distance.euclidean(point1, point2)

print(f'The Euclidean distance between {point1} and {point2} is: {euclidean_dist}')

```

**Measure Distance between point 1 and Point 2 using Cityblock distance**

```python
#Measure Distance between point 1 and Point 2 using Cityblock distance
city_block_dist = distance.cityblock(point1, point2)
print(f'The City Block (Manhattan) distance between {point1} and {point2} is: {city_block_dist}')
```

**Measure Distance between point 1 and Point 2 using Chessboard distance**

```python
#Measure Distance between point 1 and Point 2 using Chessboard distance
chessboard_dist = distance.chebyshev(point1, point2)
print(f'The Chessboard (Chebyshev) distance between {point1} and {point2} is: {chessboard_dist}')
```
